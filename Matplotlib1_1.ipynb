```python
x = [0, 2, 4, 6, 8]
y = [0, 4, 16, 36, 64]

fig,ax = plt.subplots()
ax.plot(x,y,marker='o',label = 'data_points')

ax.set_title('Basis Components of matplotlib fig')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 15)
C = np.cos(x)
S = np.sin(x)

ax = plt.axes([0.1,0.1,0.8,0.8])

ax1 = ax.plot(x,C,'bs:')

ax2 = ax.plot(x,S,'ro-')

ax.legend(labels = ('cosine','sine'),loc = 'upper left')
ax.set_title('Trigonometric Functions')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
```

```python
x = [1, 2, 3]
y = [0, 1, 0]
z = [1, 0, 1]

fig,ax = plt.subplots(2)

ax[0].plot(x,y)
ax[1].plot(x,z)
```

```python
x = np.arange(0,2.0,0.01)
y = 1 + np.sin(2 * np.pi * x)

fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)

ax1.plot(x,y,color = 'orange')
ax2.plot(x,y,color = 'blue')
ax3.plot(x,y,color = 'green')
ax4.plot(x,y,color = 'yellow')
ax5.plot(x,y,color = 'red')
ax6.plot(x,y,color = 'magenta')
```
